- Vamos a contemplar dos variables aleatorias:
    - X: variable explicativa
        - En este contexto se denomina variable de diagnóstico/prueba de diagnóstico
        - Trataremos los casos en que:
            - Sea variable de tipo Dicotómica
            - Sea variable de tipo continua
    - Y: variable respuesta
        - Recoge la presencia o ausencia de enfermedad
            - "Sí enfermo"
            - "No enfermo"/"Sano"

## Dianóstico dicotómico

- X será una variable diagnóstico dicotómica
    - X = 1 indica que la prueba diagnóstica afirma que el individuo está enfermo
    - X = 0 indica que la prueba diagnóstica dice que el individuo no está enfermo
- Y variable de enfermedad dicotómica
    - Y = 1 indica que el individuo está realmente enfermo
    - Y = 0 indica que el individuo está realmente sano
- Por tanto, el esquema es $D \leftarrow D$
- En casos de índices de riesgo con esquema $D \leftarrow D$
    - x = 1 indica que el individuo está expuesto a un factor de riesgo
    - X = 0 indica que el individuo no está expuesto a un factor de riesgo
- Por tanto, al tomar una muestra en el contexto de diagnóstico dicotómico, los datos experimentales tendrán el patrón de la siguiente tabla, donde el número de individuos:
     - Con perfil Y = 1 y X = 1 => a
     - Con perfil Y = 1 y X = 0 => b
     - Con perfil Y = 0 y X = 1 => c
     - Con perfil Y = 0 y X = 0 => d

|Y|X|
|---|---|
|1|1|
|1|0|
|...|...|
|0|0|
|0|1|

- A partir de los datos experimentales, se construye la siguiente tabla:

||X=1|X=0||
|---|---|---|---|
|Y=1|a = VP|b = FN|$r_1$|
|Y=0|c = FP|d = VN|$r_0$|
||$s_1$|$s_0$|$n$|

- Los "marginales" seran:
    - $r_1 = a + b$
    - $r_0 = c + d$
    - $s_1 = a + c$
    - $s_0 = b + d$
    
- En la tabla se indica
    - a: individuos clasificados correctamente : Verdaderos Positivos
    - b: los clasificados incorrectamente : Falsos negativos
    - c: clasificados incorrectamente : Falsos positivos
    - d: clasificados correctamente : Verdadenos negativos
    
## Sensibilidad y especifidad

- El parámetro poblacional $\theta$ sensibilidad poblacional está dado por:
<div align='center'>$\theta = se = \pi_{11}' = P(X=1|Y=1)$</div>
- Representa "el acierto de enfermos"

- El parámetro poblacional $\theta$ especifidad poblacional está dado por:
<div align='center'>$\theta = es = \pi'_{00} = P(X=0|Y=0)$</div>
- Representa "el acierto de sanos"

- Probabilidades de clasificación:
    - Fracción de verdaderos positivos (fvp)
<div align='center'>$fvp = P(X=1|Y=1) = se$</div>
    - Fracción de falsos positivos (ffp)
<div align='center'>$ffp = P(X=1|Y=0) = 1 - es$</div>

- Al ser ``se`` y ``es`` proporciones, estarán comprendidas entre 0 y 1, se expresan en % en 0% y 100%.
    - Esto nos permite utilizar IC y CH para ``se`` y ``es``

## Valor predictivo positivo y valor predictivo negativo
- El parámetro poblacional $\theta$ valor predictivo positivo poblacional ``vpp`` está dado por
<div align='center'>$\theta=vpp=\pi_{11}=P(Y=1|X=1)$</div>
- Representa el acierto en los individuos positivos según la prueba diagnóstica.


- El parámetro poblaciona $\theta$ valor predictivo negativo poblacional vpn está dado por
<div align='center'>$\theta=vnp=\pi_{00}=P(Y=0|X=0)$</div>
- Representa el acierto en los individuos negativos según la prueba diagnóstica.


- Al ser ``vpp`` y ``vpn`` proporciones estarán comprendidas entre 0 y 1 y se expresan en % entre 0% y 100%
    - Esto nos permite utilizar IC y CH para ``vpp`` y ``vpn``

## Razón de verosimilitud positiva y razón de verosimilitud negativa
- El parámetro poblacional $\theta$ de razón de verosimilitud positiva poblacional ``lrp`` viene dado por
    - lrp representa la ganacia de información cuando la prueba diagnóstica da positivo
    - Se interpretan de la siguiente forma
        - Si lrp > 1, es más que probable que un resultado poistivo en la prueba diagnóstica se dé en un sujeto enfermo que en uno sano.
<div align='center'>$\theta=lrp=\frac{\pi'_{11}}{\pi'_{10}}=\frac{P(X=1|Y=1)}{P(X=1|Y=0)}=\frac{se}{1-es}$</div>
    - El intervalo de confianza se define como:
<div align='center'>$IC(1-\alpha)\%(lrp)=\left ( exp \left ( ln \left ( \frac{\hat {se}}{1-\hat {es}} \right )\mp z_{1-\alpha/2}\sqrt{\frac{1}{r_1}\frac{1-\hat {se}}{\hat {se}}+\frac{1}{r_0}\frac{\hat {es}}{1-\hat {es}}} \right ) \right )$</div>


- El parámetro poblacional $\theta$ de razón de verosimilitud negativa poblacional ``lrn`` viene dado por
    - lrn representa la ganacia de información cuando la prueba diagnóstica da negativo
    - Se interpreta de la siguiente forma
        - lrn < 1, es menos probable que un resultado negativo en la prueba diagnóstica se dé en un sujeto enfermo que en uno sano
<div align='center'>$\theta=lrn=\frac{\pi'_{01}}{\pi'_{00}}=\frac{P(X=0|Y=1)}{P(X=0|Y=0)}=\frac{se}{1-es}$</div>
    - El intervalo de confianza se define como:
<div align='center'>$IC(1-\alpha)\%(lrn)=\left ( exp \left ( ln \left ( \frac{1-\hat {es}}{\hat {se}} \right )\mp z_{1-\alpha/2}\sqrt{\frac{1}{r_1}\frac{\hat {se}}{1-\hat {se}}+\frac{1}{r_0}\frac{1-\hat {es}}{\hat {es}}} \right ) \right )$</div>

# Diagnóstico continuo
- X es una variable de diagnóstico continua que recibe el nombre de biomarcador
- Y es una variable de enfermedad dicotómica
- El esquema es $D \leftarrow C$
- Se suele asumir
    - Valores altos de X están asociados con la presencia de enfermedad (Y=1)
    - Valores bajos de X corresponden a la ausencia de enfermedad (Y=0)
- Ejemplo, el biomarcador de cáncer de próstata PSA
    - PSA (X) sirve para diagnosticar el cáncer de próstata (Y)
- Una forma de estudiar esta situación es dicotomizar la variable X, através de una nueva $X^*$
<div align='center'>$X^*= \left \{ \begin{matrix} \mbox{1 si X} \ge c \\ \mbox{0 si X} < c \end{matrix} \right . $</div>
    - En donde c es un punto de corte o umbral que habrá que determinar de forma óptima.
- Utilizando esta $X^*$, un individuo se clasificara como
    - Positivo o "enfermo" ($X^*$=1) si el biomarcador X supera ese punto de corte
    - Negativo o "sano" en caso contrario ($X^*=0)
- Por tanto para cada c es posible definir:
    - la sensibilidad ``se(c)``, la fracción de verdaderos positivos fvp(c)
    - la especificidad ``es(c)``, la fracción de falsos positivos ffp(c)
- A partir de los valores ffp(c) y fvp(c) se define la curva roc("receiver operating characteristic").

## Curva roc
- La curva roc tiene su origen en la Teoría de la señal en los años 1950-1960, en el contexto de los radares con el objetivo de distinguir entre señal y ruido(en concreto, entre la llegada o no de misiles).
- Más adelante se introdujo en radiología con Hanley y McNeil en los años 1982 y 1983 y hoy en día es muy utilizada en Investigación clínica y también en el estudio de la bondad del comportamiento de modelos estadísticos.
- La curva roc poblacional se define como
<div align='center'>$roc(·)=\{ffp(c),fvp(c)),c\in(-\infty ,+\infty)\}$</div>
- es decir, representando para distintos puntos de corte, en el eje de abscisas la ffp(c) y en el eje de ordenadas la fvp(c). Por tanto,
<div align='center'>$roc(·)=\{ (t,roc(t),t\in(0,1)\}$</div>
    - Siendo
        - roc(t) = fvp(c) con c tal que ffp(c))=t
    - A medida que c crece, fvp(c) y ffp(c) decrecen
    - A medida que c decrece, fvp(c) y ffp(c) crecen
- El área bajo la curva roc poblacional se denota por el parámetro poblacional auc("area under the curve")
<div align='center'>$auc=\int^1_0roc(t)dt$</div>
    - Constituye un índice de resumen de la bondad del biomarcador X.
    - El auc se puede estimar por el método trapezoidal de suma de áreas de trapecios

```python
## Cálculo de la curva ROC en R

rm(list=ls())
datos=read.table(’d_c_1.txt’,header=T)
attach(datos)
library(pROC)
roc_obj <- roc(datos$rta, datos$exp)

# Imprimir el AUC
auc(roc_obj)

# Dibujar la curva ROC
## Opcion 1
plot(roc_obj, print.auc=TRUE)

## Opcion 2
roc_df <- data.frame(
fvp=rev(roc_obj\$sensitivities),
ffp=rev(1 - roc_obj\$specificities))
plot(0:10/10, 0:10/10, type=’n’, xlab="ffp", ylab="fvp")
abline(h=0:10/10, col="lightblue")
abline(v=0:10/10, col="lightblue")
abline(coef = c(0,1), col="lightblue")
with(roc_df, {
lines(ffp, fvp, type=’l’, lwd=1, col="blue")
lines(ffp, fvp, type=’b’, lwd=1, col="blue")
})
```
