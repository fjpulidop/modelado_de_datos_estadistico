## Datos
- Numéricos: medida o resultado de recuento
    - Datos discretos: elementos que pueden ser contados
    - Datos contínuos: son mediciones
- Categóricos: características
- Ordinales: escalas ordenadas

## Conjunto de datos
- Totalidad de los datos obtenidos de la muestra

## Variable
- Valor numérico variable / característica variable

## Población
- Grupo concreto de unidad de análisis

## Muestras y aleatoriedad
- Muestra es un subconjunto
- Pueden tener sesgo: favoritismo sistemático de determinadas personas o determinados resultados del estudio
- El sesgo ses evita con la aleatoriedad: las muestras individuales tienen la misma probabilidad de ser seleccionadas

## Estadístico
- Es un número que resume los datos recopilados de una muestra
    - Porcentajes, promedios, mediana o percentil

## Parámetro
- Los estadísticos se basan en datos de una muestra, no en datos de la población
- Si recopilamos los datos de la población, hacemos un censo
- Si a continuación resumimos toda la información del censo en un único número procedente de una variable, ese núymero es un parametro, no un estadístico.

## Media (promedio)
- Se usa para medir el "centro" de un conjunto
- Suma de todos los números dividido por la cantidad total

## Mediana
- Otra manera de medir el "centro" de los datos
- El valor que divide dicho conjunto en dos partes iguales, una parte con valores superiores a la mediana y otra parte con valores inferiores a la mediana

## Desviación estándar (o típica)
- "Grado de variabilidad(o dispersión)  de los números de un ocnjunto de datos."
\begin{equation}
s = \sqrt{\sum\frac{x-x^2}{n-1}}
\end{equation}

    - x=número del conjunto de datos
    - x^2 = media de los datos
    - n = número de valores del conjunto de datos

- También se utiliza para describir la zona donde deberían estar la mayoría de los datos, en relación con la media

- Si los datos siguen una distribución normal, aproximadamente el 95% de los valores se encuentran a no mas de dos desviaciones estándar de la media
- Este resultado es lo que se llama ``regla empírica``

## Percentil
- El percentil para un dato concreto es el porcentaje de valores de la muestra que están por debajo de ese dato concreto
- La mediana es el percentil 50
- Se utilizan con fines de comparación y para determinar la posición relativa

## Puntuación estándar (o típica)
- Es una manera hábil de poner en perspectiva uno de los resultados sin tener que dar un montón de detalle
- Representa el número de desviaciones estándares por encima o por debajo de la media
- El proceso de convertir un número en una puntuación estándar se llama normalización o estandarización

## Contraste de hipótesis
- Procedimiento estadístico mediante el cual se recopilan datos de una muestra y se cotejan con una afirmación referida a un parámetro poblacional
- Pruebas t: comparan dos muestras poblacionales
- Pruebas t para datos apareados: examinan diferenias entre antes y después

## Valores p
- Los contrastes de hipótesis sirven para verificar la validez de una afirmación referida a una población
- Esa afirmación que se somete a juicio se llama hipótesis nula
- La hipótesis alternativa es la que creeríamos si la hipótesis nula está equivocada
- El valor p es un número comprendido entre 0 y 1
    - Un valor p pequeño, por lo general <= 0,05 indica una prueba sólida en contra de la hipótesis nula, de manera que puedes rechazar dicha hipótesis
    - Un valor p grande (por lo general > 0,05=) indica una prueba débil en contra de la hipótesis nula, de manera que no rechazamaos dicha hipótesis
    - Valores p muy próximos al valor límite (0,05) se consideran marginales

## Coorelación y casualidad
- Coorelación: es la medida en que dos variables numéricos presentan una relación lineal
- Casualidad: un cambio en la variable x, causará un cambio en la variable y

## Cálculo de la media
\begin{equation}
\overline{x}=\frac{\sum{x_i}}{n}
\end{equation}

- $\overline{x}$ = media
- n = último número
- $x_i$ = x es cada valor del conjunto de datos. $_i$ es índice de 1 a n

- La media de una población entera se designa con la letra $\mu$ y se llama media poblacional

## Cálculo de la mediana
1. Orden de menor a mayor
2. Si es impar -> eleigmos la del centro
3. Si es par -> se cogen las 2 centrales y se calcula la media

## Cálculo de la desviación estándar
\begin{equation}
s = \sqrt{\frac{\sum(x - \overline{x})^2}{n-1}}
\end{equation}

en dónde, recordemos:

\begin{equation}
\frac{\sum(x - \overline{x})^2}{n-1} 
\end{equation}

es la variacia muestral (s^2).
- La desviación estándar de una población entera se designa con la letra $\sigma$
- Variación estándar pequeña = la mayoría de los valores están próximos al centro (media)

## Fuera de rango
- Rango es la diferencia entre el valor más grande y el valor más pequeño del conjunto de datos
- Rango intercuartílico: rango del 50% de los valores centrales

## Regla empírica
- Si es una población tiene una distribución normal, con una media poblacional $\mu$ y una desviación estándar $\sigma$
    - 68% de los valores se encuentran a 1 $\sigma$ de la media. $\mu\pm\sigma$
    - Aprox 95% se encuentran a 2 $\sigma$ de la media. $\mu\pm2\sigma$
    - Aprox 99,7% se encuentran a 3 $\sigma$ de la media. $\mu\pm3\sigma$

## Posición relativa con percentiles
- Percentil K es un número del conjunto de datos que divide los datos en dos partes:
    - La parte inferior contiene K% de los datos
    - La parte superior contiene el resto de los datos
- El percentil es tu posición con respecto al resto
    - Top 15% 
    
## Rango intercuartílico
- RIC = Q3 - Q1 = 75% - 25% = 50%
- RIC es mejor que rango normal => descarta outliers
