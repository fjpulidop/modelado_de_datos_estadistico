## Variables aleatorias y distribución binomial
- Una v.a. es una característica, una medida o una cifra de recuento que cambia aleatoriamente dentro de un determinado conjunto o con arreglo a una determinada pauta.
- Su notación es X, Y, Z ...

## V.a. discretas
- Si se pueden enumerar en números enteros 0,1,2... 
    - Finitas [0,1,2]
    - Infinitas [0,1,...]

## V.a. continuas
- Solo se pueden describir utilizando un intervalo de números reales
- Son infinitos no numerables

## Distribución de probabilidad
- Una v.a. discreta tiene un conjunto de valores posibles, cada uno de los cuales tiene unaprobablidad de ocurrir
- La notación empleada para cualquier resultado específico es x
- Un dado {1,2,3,4,5,6} -> p(x) -> probablidad de que ocurra cualquier resultado específico
- Probabilidad de sacar un 1 -> p(1)
- Ejemplo de los inquilinos y los perros
    - 0,4% tiene 1 perro
    - 0,5% no tiene perro
    - 0,07% tiene 2 perros
    - 0,03% tiene 3 perros

|x|p(x)|
|---|---|
|0|0,5|
|1|0,4|
|2|0,07|
|3|0,03|

## Media y variancia de una v.a. discreta
- La ``media de una v.a.`` es el promedio de todos los resultados que cabe esperar a largo plazo (para todas las muestras posibles)
- Notación: $\mu_x$

- La ``variancia de una v.a.`` es el promedio de los cuadrados de las desviadciones de todos los resultados que obtendrías a largo plazo respecto a la media aritmética de la población, para cada una de las muestras posibles.
- Notación: $\sigma_{x^2}$

## Desviación estandar de v.a.
- La desviación estandar de una v.a. es la raíz cuadrada de la variancia y se designa como $\sigma_x$
- Mas o menos representa el promedio de la desviación respecto de la media

## V.a. Binomial
- Se caracteriza por
    - Nº fijo de intentos (n)
    - Cada intento tiene 2 resultados posibles: éxito/fracaso
    - La probabilidad de éxito (la llamaremos p) es la misma en cada intento
    - Los intentos son independientes, es decir, el resultado de otro intento no influye
- p: probabilidad de obtener un resultado de éxito
- p(x): probabilidad de obtener x éxitos en n intentos

## Encontrar probabilidades binomiales utilizando una fórmula:
\begin{equation}
\binom{n}{x}p^x(1-p)^{n-x}
\end{equation}

en donde:
    - n: número de intentos
    - x: número de éxitos
    - $\binom{n}{x}$: nº de maneras en que puede obtenerse x éxitos en n intentos
    - p^x: probabilidad de éxito en un intento
    - 1-p: probabilidad de fracaso en un intento (también llamado ``q``)
    
iugualmente podemos decir que:
\begin{equation}
\binom{n}{x}=\frac{n!}{x!(n-x)!}
\end{equation}

- La notación n! ( n factional o factorial de n) indica el número de maneras de reordenar n elementos
- Para calcular n! hay que multiplicar n(n-1)(n-2)..(2)(1)
- Por ejemplo, 5! es 5(4)(3)(2)(1)=120;
- 1! es 1
- Por convención, 0! es 1

## Ejercicio de ejemplo de v.a. binomial
- Supongamos que tenemos que pasar por 3 semáforos para llegar a tu lugar de trabajo. Supongamos que X es el número de semáforos que encuentras en rojo. ¿Cuántas maneras hay de que te encuentres dos semáforos en rojo de camino al trabajo?
- (omitimos naranja) Siendo V=verde y R=rojo, por tanto existen las isguientes maneras: VRR, RVR y RRV
- 1 intento = 1 semáforo
- 1 éxito = 1R
- n=3
- x=2
- Por tanto:
\begin{equation}
\binom{3}{2} = \frac{3!}{2!(3-2)!}=\frac{3(2)(1)}{[(2)(1)](1)}=\frac{6}{2}=3
\end{equation}


- Ahora, suponiendo que los semáforos son independientes entre sí y que cada uno tiene una probabilidad de 30% de estar en rojo. Imagina que quieres encontrar la distribución de probabilidad par X, es decir, una lista de todos los valores posibles de X(0,1,2,3) y sus probabilidades:
- n= 3 intentos
- X=nº de semáforos en rojo tiene una distribución binomial
- Si vamos a las fórmulas, 1-p = probabilidad de un semáforo NO rojo = 1 - 0,3 = 0,7; y el nº de semáforos no rojos es de 3-X
- Aplicando la fórmula de p(x) se obtienen las siguientes probabilidades par x=0,1,2 y 3 semáforos en rojo.
\begin{equation}
p(0)=\binom{3}{0}0.3^0(1-0,3)^{3-0} = \binom{3!}{0!(3-0)!}(0,3)^0(0,7)^3 = 1(1)(0,343)= 0,343;
p(1)=\binom{3}{1}0.3^1(1-0,3)^{3-1} = \binom{3!}{1!(3-1)!}(0,3)^1(0,7)^2 = 3(0,3)(0,49)= 0,441;
p(2)=\binom{3}{2}0.3^2(1-0,3)^{3-2} = \binom{3!}{2!(3-2)!}(0,3)^2(0,7) = 3(0,09)(0,70)= 0,189;
p(3)=\binom{3}{3}0.3^3(1-0,3)^{3-3} = \binom{3!}{3!(3-3)!}(0,3)^3(0,7)^0 = 1(0,027)(1)= 0,027;
\end{equation}
- Distribución binomial para X=nº de semáforos en rojo:
    - n=3
    - p=0,3
    
|x|p(x)|
|---|---|
|0|0,343|
|1|0,441|
|2|0,189|
|3|0,027|

- La suma es 1
- Se suelen utilizar minitablas ya precargadas con las distribuciones binomiales dependientes de n y p.

## Encontrar probabilidades para X mayor que un valor, menor que un valor o entre dos valores
- Basta con buscar en la tabla y sumar sus probabilidades
- Siguiendo el ejemplo de los semáforos si quieres conocer la probabilidad de encontrarte con más de un smeáforo en rojo, p(x-1) tiene que sumar p(2) + p(3) de la tabla binomial.

- Como máximo: $\le$x

## Comprobar la media y la desviación estándar de la distribución binomial
- Si X tiene una distribución binomial con n intentos y una probabilidad de éxito p en cada intento, entonces:
    1. La media de X es $\mu$=np
    2. La variancia de X es $\sigma^2$=np(1-p)
    3. La desviación estándar de X es $\sigma$=$\sqrt{np(1-p)}$
- Por ejemplo, pongamos que lanzas una moneda al aire 100 veces y X es el número de caras. En ese caso X tiene una distribución binomial en la que n=100 y p=0,5. La media de $\mu$=np=100(0,5) caras
- La variancia de X es $\sigma^2$=np(1-np)=100(0,5)(1-(100(0,5))=25, que está en unidades cuadradas (por lo que no se puede interpretar).
- La desviación estándar es la raíz cuadrada de la variancia,osea, 5.
- Esto significa que si lanzamois 100 monedas, el nº medio de caras es 50, y cabe esperar que esta cifra varíe 5 caras de promedio.
