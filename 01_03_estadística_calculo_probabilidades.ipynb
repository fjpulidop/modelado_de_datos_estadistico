## Cálculo de probabilidades
- El cálculo de probabilidades es la parte más matemática de la Estadística y suministra las herramientas (por ejemplo, el teorema central del límite) para poder conocer la población a partir de una muestra representativa de ella (para poder pasar a lo general partiendo de lo concreto o particular).


- En este proceso es clave el concepto de parámetro y de distrivución muestral de un estadístico.


- En la Estadística clásica (también llamada frecuentista) se está interesado en conocer una población a través de un parámetro poblacional $\phi$ que es una constante desconocida y que se desea saber su valor.


- Si se está interesado en conocer el impacto de una enfermedad en una población, sería interesante conocer el parámetro proporcion poblacional  de dicha enfermedad ( en este caso, también denominada `` prevalencia poblacional de enfermedad``), que se podría denotar por la proporción poblacional $\pi$ y que representa la probabilidad poblacional de enfermedad, es decir $\pi = P(Y = 1)$).


- Para conocer $\pi$ (siempre que se pueda, existe el convenio de denotar con letras griegas a los parámetros poblacionales) se cuenta con la ayuda de la proporción muestral, que se podría denotar por $\hat\pi$ dado $\hat\pi=\frac{r_1}{n}$ (en otros textos se denota a la proporción muestral por $p$ para hacer enfasis en que se utilizan letras latinas para los valores muestrales; cuando ponemos el ``^`` es para indicar que es un valor estimado de $\pi$. 

- La media téorica (esperanza matemática) de Y => E[Y]=$\pi$


- La varianza teóriza de Y => V[Y]=$\pi(1-\pi)$


- La desviación típica de Y => D[Y]=$\sqrt{\pi(1-\pi)}$


- Ahora se trata de extrapolar de la muestra a la población. Se consigue a través del concepto de distribución muestral de un estadístico. 


- La proporción muestral tiene un determinado valor, por ejemplo 0.34. Si se hubiese tomado otra muestra representativa, la proporción muestral tomaría seguramente otro valor, por ejemplo 0.56 y si se hubiese tomado otra, quizás 0.52, etc. Si este proceso de tomar muestras representativas se repitiese varias veces, se podría contruir una columna de datos con los valores de todas las proporciones muestrales obtenidas, parecidas a la siguiente tabla, donde dichos valores varían de unos a otros (hay una cierta variabilidad).

- |$\hat\prod$| indica que se está ante una nueva variable aleatoria (de ahí la mayúscula), que es la variable aleatoria asociada al estadístico proporción muestral.


- Esta variable tendrá un histograma asociado (un modelo de probabilidad continuo asociado).


- Este histograma o modelo asociado, recibe el nombre de distribución muestral del estadístico proporción muestral. En este caso el teorema central del límite afirma que tiene forma de campana de Gauss (es una v.a. Normal). Desde un punto de vista formal si se considera $Y_i$ la v.a. $Y$ asociada al individu o $i$, se tiene que el estadístico $\hat\theta$ dado por la v.a. proporción muestral $\hat\prod$ es
<div align="center">$\hat\theta=\hat\prod=\hat Y=\frac{1}{n}\sum^n_{i=1}Y_i$</div>


- Con lo que la media teórica de $E[\hat\prod]=\pi$


- Y al ser $Y_i$ independientes, son incorreladas, por lo que la varianza teórica de $\hat\prod=\hat Y$, esto es $V[\hat\prod]=V[\hat Y] = \pi(1-\pi)\frac{1}{n}$


- La raíz cuadrática de la varianza teórica del estadístico proporción muestral $\hat\prod$ se denomina error estándar del estadístico proporción muestral, $EE[\hat\prod]$ y representa una medida de la dispersión del estadístico proporción muestral. Su expresión viene dada por:
<div align="center">$EE[\hat\prod] = \sqrt{V[\hat\prod]} = \sqrt{\pi(1-\pi)\frac{1}{n}}$</div>
    
    
- Con estos elementos y la ayuda de la estadística inferencial se puede inferir qué ocurre con el parámetro proporcional $\pi$, dando el salto de la muestra a la población. Dentro de la estadística inferencial se distingue la inferencia mediante un intervalo de confianza (IC) y la inferencia mediante un contraste de hipótesis (CH)
