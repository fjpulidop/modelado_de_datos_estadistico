# Índices de riesgo

- Hay que tener conocimiento previo de contraste de hipótesis e intervalo de confianza.


- Tendremos datos tipo esto:
    - Y: es la variable ``respuesta`` (dependiente, resultado u objetivo)
    - X: es la variable ``explicativa`` (independiente o factor de exposición)
    
    
- Escribimos $D \leftarrow D$ para indicar que estamos ante un modelo estadístico que establece que se está intentando explicar una variable dicotómica a través de otra variable dicotómica.


- El patrón será el siguiente:

|Y|X|
|---|---|
|1|1|
|0|1|
|...|...|
|0|0|
|1|0|


- Donde podremos construir la siguiente tabla:

||X=1|X=0||
|---|---|---|---|
|Y=1|a|b|$r_1$|
|Y=0|c|d|$r_0$|
||$s_1$|$s_0$|$n$|


- Los "marginales" seran:
    - $r_1 = a + b$
    - $r_0 = c + d$
    - $s_1 = a + c$
    - $s_0 = b + d$


- El tamaño de la muestra será: $n$


- El diseño del estudio fija los marginales:
    - Diseño prospectivo (hacia adelante): 
        - se fija $s_1$ y $s_0$
        - se pueden utilizar como índices de riesgo:
            - Diferencia de riesgos
            - Riesgo relativo
            - Odds ratio
    - Diseño retrospectivo (hacia atrás):
        - se fija $r_1$ y $r_0$
        - se pueden utilizar como índices de riesgo:
            - Odds ratio
    - Diseño transversal:
        - se fija $n$
        - se pueden utilizar como índices de riesgo:
            - Odds ratio


- Índice de riegos absoluto = diferencia de riesgos


- Índice de riesgo relativo = riesgo relativo y odds ratio


## Índices Absolutos

### Diferencias de riesgo

- Con intervalo de confianza $(1-\alpha)\%$ la diferencia de riesgo se define como:

\begin{equation}
IC(1-\alpha)\%(dr)=({\frac{a}{s_1}-\frac{b}{s_0}\mp z_{1-{\alpha/2}}\sqrt{\frac{ac}{s^3_1}+\frac{bd}{s^3_0}}})
\end{equation}


- En lo que respecta a la hipótesis:
    - $H_0: dr=0$
    - $H_1: dr\ne0$
    
    
- Hay que realizarlo a través del test $z$ de diferencia de proporciones, de Fisher o del test $\chi^2$


## Índices relativos
### Riesgo relativo

- Con intervalo de confianza $(1-\alpha)\%$ el riesgo relativo se define como:

\begin{equation}
IC(1-\alpha)\%(rr)= \left ( exp \left ( ln \left ( \frac{a / s_1}{b / s_0} \right ) \mp z_{1-\alpha/2} \sqrt{\frac{1}{a}-\frac{1}{s_1}+\frac{1}{b}-\frac{1}{s_0} } \right ) \right )
\end{equation}


- En lo que respecta al contraste de hipótesis:
    - $H_0: rr=1$
    - $H_1: rr\ne1$
    

- dado que $rr = 1 \leftrightarrow dr = 0$ hay que realizarlo a través del test $z$ de diferencia de proporciones, de Fisher o del test $\chi^2$.

### Odds ratio

- Con intervalo de confianza $(1-\alpha)\%$ el riesgo relativo se define como:

\begin{equation}
IC(1-\alpha)\%(or)= \left ( exp \left ( ln \left ( \frac{ad}{b / bc} \right ) \mp z_{1-\alpha/2} \sqrt{\frac{1}{a}+\frac{1}{b}+\frac{1}{c}+\frac{1}{d} } \right ) \right )
\end{equation}


- En lo que respecta al contraste de hipótesis:
    - $H_0: or=1$
    - $H_1: or\ne1$
